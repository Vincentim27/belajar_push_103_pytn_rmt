## B. Data Visualization (matplotlib)

```python
# Contoh jelek ketika menggunakan .plot nya pandas:
haiti = df.loc['Haiti',years]
haiti.plot()
```

```python
# Sekarang kita revamp pakai matplotlib
haiti.index = haiti.index.map(int)
haiti.plot()

plt.title('Immigration from Haiti')
plt.xlabel('Number of Immigrants')
plt.ylabel('Years')
```

```python
# Berikan anotasi/catatan di lonjakan tahun 2010 ketika Kanada meningkatkan upaya penerimaan pengungsi
haiti.plot()

plt.title('Immigration from Haiti')
plt.xlabel('Number of Immigrants')
plt.ylabel('Years')
plt.text(2005,6000,'2010 Earthquake')
```

Contoh soal 1: Compare immigrants India and China from 1980 to 2013

```python
# Contoh salah, karena value nya jadi tahun, x nya jadi negara
df.loc[['China','India'], years].plot()
```

```python
# Gunakan transpose untuk switch x dan y.
df.loc[['China','India'], years].T.plot()
```

Contoh soal 2: Compare the trend of top 5 countries that contributed the most to immigration to Canada

```python
df.sort_values('Total', ascending=False).head()[years].T.plot(figsize=(14,8))
plt.title('Immigration Trend of Top 5 Countries')
plt.xlabel('Numer of Immigrants')
plt.ylabel('Years')

plt.show()
```

Cosol 3: Compare the number of Iceland immigrants (country = 'Iceland') to Canada from 1980 to 2013

```python
df.loc['Iceland',years].plot(kind='barh', figsize=(12,10))

plt.title('Iceland Immigrants to Canada from 1980 to 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.show()
```
