#### ***Study Case***: asumsinya belum ada kolo game result / akhirnya kita mau bandingin antara kondition kita dengan data

```python
df_ori.loc[df_ori.team_id == 'LAL', ['team_id','pts','opp_pts']]
```

```python
df_ori.game_result.unique()
```

```python
list_kosong = []

for i in df_ori.team_id:
    # print(i)
    if i == 'LAL':
       
      list_kosong.append('LAL')
    else:
      list_kosong.append('not_LAL')

   

df_ori['kolom_baru'] = list_kosong
df_ori.head()
```

```python
# buat kolom baru namanya 'new_column'
df_ori['new_column'] = df_ori.pts - df_ori.opp_pts
```

```python
# tampilkan kolom 'new_column' 
df_ori.new_column
```

```python
# buat kolom baru lagi namanya 'jadi menang atau gak nich' dgn kondisi positif/ negatif
df_ori['jadi_menang_atau_kalah_nich']= np.where(df_ori.new_column > 0,'W','L')
```

```python
# cek data asli
df_ori.loc[df_ori.team_id == 'LAL'].groupby(['game_result'])[['game_result']].value_counts()
```

```python
# cek data baru dari kolom baru
df_ori.loc[df_ori.team_id == 'LAL'].groupby(['jadi_menang_atau_kalah_nich'])[['jadi_menang_atau_kalah_nich']].value_counts()
```

berarti kesimpulannya kolom baru yg kita bikin dgn kondisi yg kitabikin sudah betul/tepat karena sesuai dengan data asli based on game_result column.

Create new data dengan kondisi di pandas version 2.0 ke atas append sudah tdk bisa digunakan, alternatif nya menggunakan `pd.concat`

```python
# membuat baris (data) baru
data_baru_nie = {
    'gameorder': 70000, 
    'game_id': None, 
    'lg_id': None, 
    '_iscopy': None, 
    'year_id': None, 
    'date_game': None,
    'seasongame': None, 
    'is_playoffs': None, 
    'team_id': None, 
    'fran_id': None, 
    'pts': None, 
    'elo_i': None,
    'elo_n': None, 
    'win_equiv': None, 
    'opp_id': None, 
    'opp_fran': None, 
    'opp_pts': None, 
    'opp_elo_i': None,
    'opp_elo_n': None, 
    'game_location': None, 
    'game_result': None, 
    'forecast': None, 
    'notes': None,
    'kolom_baru': None, 
    'new_column': None, 
    'jadi_menang_atau_kalah_nich': None
}
df_ori = pd.concat([df_ori,pd.DataFrame([data_baru_nie])], ignore_index=True)
df_ori.tail()
```
